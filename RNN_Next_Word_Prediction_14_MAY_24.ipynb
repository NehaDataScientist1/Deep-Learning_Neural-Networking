### NEXT WORD PREDICTION USING SEQ2SEQ MODEL TENSORFLOW, KERAS, LSTM

```python
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
df= pd.read_csv(r'/content/tmdb_5000_movies.csv')
#r'C:\Users\Neha\Downloads\NARESH IT DATA SCIENCE\MAY\1- 18 MAY\14 MAY\Next Word Predictor_ LSTM_SEQ\tmdb_5000_movies.csv')
```

```python
df.head(3)
```

```python
df = df['original_title']
```

```python
df
```

```python
movie_name = df.to_list()
```

```python
movie_name
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer()
```

```python
tokenizer
```

```python
tokenizer.fit_on_texts(movie_name)
```

```python
tokenizer
```

```python
seq = tokenizer.texts_to_sequences(movie_name)
```

```python
seq
```

```python
seq[:10]
```

```python
tokenizer.index_word
```

```python
X = []
Y = []
total_words_dropped = 0

for i in seq:
  if len(i) > 1:
    for index in range(1, len(i)):
      X.append(i[:index])
      Y.append(i[index])
  else:
    total_words_dropped += 1
```

```python
print("Total Single Words Dropped are:", total_words_dropped)
```

```python
X[:10]
```

```python
Y[:10]
```

```python
X= tf.keras.preprocessing.sequence.pad_sequences(X)
```

```python
X
```

```python
X.shape
```

```python
Y= tf.keras.utils.to_categorical(Y)
```

```python
Y
```

```python
Y.shape
```

```python
vocab_size= len(tokenizer.index_word) + 1
```

```python
vocab_size
```

#Reccurent Neural Networking Model:

```python
model = tf.keras.Sequential([
  tf.keras.layers.Embedding(vocab_size, 14  ),
    tf.keras.layers.LSTM(100, return_sequences= True),
    tf.keras.layers.LSTM(100),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(Y.shape[1], activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer= tf.keras.optimizers.Adam(learning_rate=0.004), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X, Y, batch_size=200, epochs=250)
```

```python
model.save('neha_NWP.h5')
```

```python
vocab_array= np.array(list(tokenizer.word_index.keys()))
```

```python
vocab_array
```

```python
def make_prediction(text, n_words):
    for i in range(n_words):
        text_tokenize = tokenizer.texts_to_sequences([text])
        text_padded = tf.keras.preprocessing.sequence.pad_sequences(text_tokenize, maxlen=14)
        prediction = np.squeeze(np.argmax(model.predict(text_padded), axis=-1))
        prediction = str(vocab_array[prediction - 1])
        print(vocab_array[np.argsort(model.predict(text_padded)) - 1].ravel()[:-3])
        text += " " + prediction
    return text
```
